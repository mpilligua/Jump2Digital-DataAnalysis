#### Merging datasets

```python
alquiler_data = pd.read_csv('Precio_alquiler_m2_final.csv', sep=';')
alquiler_data
```

```python
# export to csv
alquiler_data.to_csv('Precio_alquiler_m2_final_2.csv', index=False, decimal= ",", sep = ';')
```

```python
alquiler_data = alquiler_data[["Barri_cod", "Mitjana", "Màxim", "Minim", "Preu_mitja_local"]]
# alquiler_data['Barri_cod'] = alquiler_data['Barri_cod'].astype(int)

rename = {"Barri_cod": "Codi_Barri", "Mitjana": "Mitjana_alquiler", "Màxim": "Maxim_alquiler", "Minim": "Minim_alquiler", "Preu_mitja_local": "Mitjana_alquiler_local"}
alquiler_data = alquiler_data.rename(columns=rename)
alquiler_data.head()
```

```python
atur_data = pd.read_csv('2019_Durada_atur.csv', sep=',')
atur_data
```

```python
cols2keep = ["Codi_Barri", "Durada_atur", "Nombre"]
atur_data = atur_data[cols2keep]
rename = {"Nombre": "Nombre_aturs"}
atur_data.rename(columns=rename, inplace=True)
atur_data.head()
```

```python
# merge it to the principal dataframe
merged_data = pd.merge(alquiler_data, atur_data, left_on='Codi_Barri', right_on='Codi_Barri')
merged_data.head()
```

```python
# export to csv
merged_data.to_csv('merged_data.csv', index=False, encoding='latin-1')
merged_data
```

```python
geodata = census_data_common[columns2keep]
geodata.head()
```

```python
# use geopanda to plot the data in a map
import geopandas as gpd
import matplotlib.pyplot as plt

geodata = geodata[geodata["Nom_Sector_Activitat"]=="Comerç al detall"]
# geodata = geodata[geodata["Nom_Grup_Activitat"]=="Equipament personal"]
geodata = geodata[geodata["Nom_Grup_Activitat"] == "Quotidià alimentari"]


# create a geodataframe
gdf = gpd.GeoDataFrame(geodata, geometry=gpd.points_from_xy(geodata.Longitud, geodata.Latitud))

# load the map of Barcelona
bcn_map = gpd.read_file('barris.geojson')

# plot the data with the map of Barcelona
fig, ax = plt.subplots(figsize=(15,15))
bcn_map.plot(ax=ax, alpha=0.4, color='grey')
gdf.plot(ax=ax, markersize=2, color='green')
ax.set_axis_off()
# set background color to none
fig.set_facecolor('none')
plt.show()
```
